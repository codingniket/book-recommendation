```python
import pandas as pd

books = pd.read_csv('books.csv')

books.head()
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()

sns.heatmap(books.isna().transpose(), cbar = False , ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")
plt.show()
```

```python
books["categories"].value_counts().reset_index().sort_values("count",ascending=False)
```

```python
books.loc[:, "words_in_description"] = books["description"].str.split().str.len()
```

```python
# books.loc[books["words_in_description"].between(1,4),"description"]
books
```

```python
book_missing_25_words = books[books["words_in_description"] >= 25]
```

```python

book_missing_25_words
```

```python
book_missing_25_words.loc[:,"tagged_description"] = (book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1))
```

```python
book_missing_25_words
```

```python

(
    book_missing_25_words
    .drop(["words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index = False)
)
```
