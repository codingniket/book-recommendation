```python

from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_chroma import Chroma
```

```python

from dotenv import load_dotenv

load_dotenv()
```

```python
import pandas as pd

books = pd.read_csv("books_cleaned.csv")
```

```python
books
```

```python
books["tagged_description"]
```

```python
books["tagged_description"].to_csv("tagged_description.txt", index=False, header=False, lineterminator ='\n')
```

```python
raw_documents = TextLoader("tagged_description.txt",encoding="utf-8").load()
text_splitter = CharacterTextSplitter(chunk_size=1, chunk_overlap=0, separator="\n")

documents = text_splitter.split_documents(raw_documents)
```

```python
documents[5]
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embedder = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

db_books = Chroma.from_documents(
    documents,
    embedding=embedder
)
```

```python
query = "A Children book on nature"
docs = db_books.similarity_search(query, k = 5)
docs
```

```python
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 5,
) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k = 5)

    books_list = []

    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]

    return books[books["isbn13"].isin(books_list)]
```

```python
retrieve_semantic_recommendations("A children book on nature")
```
